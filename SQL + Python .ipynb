```python
#extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip') 
zip_ref.extractall() # extract file to dir
zip_ref.close() # close file
```

```python
#read data from the file and handle null values
import pandas as pd
import numpy as np
df = pd.read_csv('orders.csv')
df.head(10)
```

```python
# check for null values in the data
df.info()
```

```python
# check for null values in Ship Mode column
df['Ship Mode'].unique()
```

```python
# Replacing all instances of 'Not Available' and 'unknown' with np.nan
df['Ship Mode'].replace({'Not Available': np.nan, 'unknown': np.nan}, inplace=True)

# inplace=True parameter tells Pandas to modify the DataFrame in place, rather than creating a new copy
```

```python
# check if the null values in Ship Mode column are replaced with np.nan
df['Ship Mode'].unique()
```

```python
#convert order date from object data type to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d')
```

```python
# Confirm the conversion
df.info()
```

```python
# Convert Column names to lower case with no spaces

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#derive new columns discount , sale price and profit

df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']= df['list_price']-df['discount']
df['profit']=((df['sale_price']*df['quantity'])-df['cost_price'])
df.head(5)
```

```python
#drop cost price list price and discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
